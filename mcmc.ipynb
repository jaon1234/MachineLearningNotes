```python
import numpy as np
```

```python
# 协方差矩阵
cov = [[1,0.5],[0.5,1]]
# 均值
mean = [1,1]
data = np.random.multivariate_normal(mean,cov,500,'raise')
import matplotlib.pyplot as plt
plt.scatter(data[:, 0], data[:, 1])
plt.xlim(-3, 5)
plt.ylim(-3, 5)
plt.show()
```

```python
## 下面是用Gibbs算法进行模拟
# 1. 初始化
x_i_1 = [100,100]
data = []
data.append(x_i_1)
# 按样本量迭代，每次迭代分别抽取2个分量
for i in range(5):
    # 为第i个样本创建局部变量
    x_i = []
    # 抽取第一个分量
    x_i.append(np.random.normal(0.5*x_i_1[1],0.5*0.5))
    # 抽取第二个分量
    x_i.append(np.random.normal(0.5*x_i[0],0.5*0.5))
    # 将抽取的样本添加到样本集中
    data.append(x_i)
    # 将x_i设置为x_i-1
    x_i_1 = x_i
```

```python
data = np.array(data)
plt.scatter(data[:,0],data[:,1])
# plt.xlim(-3, 5)
# plt.ylim(-3, 5) 
plt.show()
```
